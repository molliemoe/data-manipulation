## Getting Started

Import the CSVs for each of the following cities: Baltimore, Boston, New York, and Philadelphia. Set up a dataframe for each city.

```python
# Import the necessary libraries and CSVs. Make some dataframes!
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

baltimore_df = pd.read_csv("baltimore_9-24-2016_9-30-2017.csv")
boston_df = pd.read_csv("boston_9-24-2016_9-30-2017.csv")
new_york_df = pd.read_csv("new-york_9-24-2016_9-30-2017.csv")
philadelphia_df = pd.read_csv("philadelphia_9-24-2016_9-30-2017.csv")
```

## Clean Your Data

In the last lesson, we cleaned the data to related to San Francisco. Pull up your notebook from the last lesson and use your cleaning skills to clean the dataframes as necessary.

```python
# Clean your data here!
# Baltimore Data Clean

#Drop empty columns
baltimore_df = baltimore_df.drop(["Origin District", "Environment", "Unit of Sale", "Quality", "Condition", "Appearance", "Storage", "Crop", "Trans Mode", "Package", "Repack", "Commodity Name", "City Name", "Grade", "Mostly Low", "Mostly High"], axis=1)

# Fill the Missing Values in the Type Column
baltimore_df["Type"] = baltimore_df["Type"].fillna("Conventional")

#Drop Duplicates
baltimore_df = baltimore_df.drop_duplicates()

#Check status
baltimore_df.head()
```

```python
# Boston Data Clean

#Drop empty columns
boston_df = boston_df.drop(["Origin District", "Environment", "Unit of Sale", "Quality", "Condition", "Appearance", "Storage", "Crop", "Trans Mode", "Package", "Repack", "Commodity Name", "City Name", "Grade", "Mostly Low", "Mostly High"], axis=1)

# Fill the Missing Values in the Type Column
boston_df["Type"] = boston_df["Type"].fillna("Conventional")

#Drop Duplicates
boston_df = boston_df.drop_duplicates()

#Check status
boston_df.head()
```

```python
# New York Data Clean

#Drop empty columns
new_york_df = new_york_df.drop(["Origin District", "Environment", "Unit of Sale", "Quality", "Condition", "Appearance", "Storage", "Crop", "Trans Mode", "Package", "Repack", "Commodity Name", "City Name", "Grade", "Mostly Low", "Mostly High"], axis=1)

# Fill the Missing Values in the Type Column
new_york_df["Type"] = new_york_df["Type"].fillna("Conventional")

#Drop Duplicates
new_york_df = new_york_df.drop_duplicates()

#Check status
new_york_df.head()
```

```python
# Philadelphia Data Clean

#Drop empty columns
philadelphia_df = philadelphia_df.drop(["Origin District", "Environment", "Unit of Sale", "Quality", "Condition", "Appearance", "Storage", "Crop", "Trans Mode", "Package", "Repack", "Commodity Name", "City Name", "Grade", "Mostly Low", "Mostly High"], axis=1)

# Fill the Missing Values in the Type Column
philadelphia_df["Type"] = philadelphia_df["Type"].fillna("Conventional")

#Drop Duplicates
philadelphia_df = philadelphia_df.drop_duplicates()

#Check status
philadelphia_df.head()
```

## Combine Your Data

Now that you have four clean sets of data, combine all four into one dataframe that represents the entire Northeast region.

```python
# Combine the four dataframes into one!

northeast_df = pd.concat([baltimore_df, boston_df])
northeast_df = pd.concat([northeast_df, new_york_df])
northeast_df = pd.concat([northeast_df, philadelphia_df])

#Check Status
northeast_df.head()
```
